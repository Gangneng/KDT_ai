## 2차함수 그래프 with plot()

```python
# numpy.array를 이용해 함수 그래프 그리기
# y = x^2
x = np.array([1, 2, 3, 4, 5])# 정의역
y = np.array([1, 4, 9, 16, 25])# f(x)

plt.plot(x,y)
plt.show()
```

```python
# np.arange(a, b, c) range에는 c부분에 정수만 넣을 수 있는데, np.arange와 같은 경우는 실수형도 넣을 수 있다.

x = np.arange(-10, 10, 0.01)

plt.plot(x,x**2)
plt.show()
```

```python
# x, y축에 설명 추가하기

x = np.arange(-10, 10, 0.01)

plt.xlabel('x')
plt.ylabel('f(x) value')

plt.plot(x,x**2)
plt.show()
```

```python
# x, y축 범위 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel('x')
plt.ylabel('f(x) value')

plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]

plt.plot(x,x**2)
plt.show()
```

```python
# x축, y축에 눈금 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel('x')
plt.ylabel('f(x) value')

plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]

plt.xticks([i for i in range(-5, 5, 1)])
plt.yticks([i for i in range(0, 24, 3)])

plt.plot(x,x**2)
plt.show()
```

```python
# 그래프에 title 달기

x = np.arange(-10, 10, 0.01)

plt.xlabel('x')
plt.ylabel('f(x) value')

plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]

plt.xticks([i for i in range(-5, 5, 1)])
plt.yticks([i for i in range(0, 24, 3)])

plt.title(" y = x^2 graph ")

plt.plot(x,x**2)
plt.show()
```

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel('x')
plt.ylabel('f(x) value')

plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]

plt.xticks([i for i in range(-5, 5, 1)])
plt.yticks([i for i in range(0, 24, 3)])

plt.title(" y = x^2 graph ")

plt.plot(x,x**2, label="trend")
plt.legend()

plt.show()
```

# 3. Matplotlib Case Study

## 꺾은선 그래프 (Plot)

`.plot()`

```python
x = np.arange(20)
y =  np.random.randint(0, 20, 20) #난수를 20번 생성

plt.plot(x, y)
plt.show()
```

```python
# Extra : y축을 20까지 보이게 하고 싶다면, y축을 5단위로 보이게 하고 싶다면?

x = np.arange(20)
y =  np.random.randint(0, 20, 20) #난수를 20번 생성

plt.axis([0, 20, 0, 20])

#plt.yticks([i for i in range(0,21,5)])
plt.yticks([0, 5, 10, 15, 20])

plt.plot(x, y)
plt.show()
```

## 산점도 (Scatter Plot)

`.scatter()`

```python
plt.scatter(x, y)
plt.show()
```

## 박스 그림 (Box Plot)

- 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max)

```python
plt.boxplot(y)

plt.show()
```

```python
# Extra : Plot의 title을 "Box plot of y"

plt.title("Box plot of x, y")

plt.boxplot((x, y))

plt.show()
```

## 막대 그래프 (Bar Plot)

- 범주형 데이터의 값과 그 값의 크기를 직사각형으로 나타낸 그림

```python
plt.bar(x, y)
plt.show()
```

```python
# Extra : xtics를 올바르게 처리해보자.

plt.xticks(np.arange(0,20,1))

plt.bar(x, y)
plt.show()
```

```python
# cf) Histogram
# 공통점 - 도수분포를 직사각형의 막대 형태로 나타낸다
# 차이점 - "계급"으로 나타낸 것이 특징이다. 0, 1, 2가 아니라 0~2까지의 "범주형" 데이터로 구성 후 그림을 그림

plt.xticks(np.arange(0,20,2))

plt.hist(y, bins=np.arange(0, 22, 2))

plt.show()
```

```python
z = [100,300,200,400]

plt.pie(z, labels=['one','two','three','four'])
plt.show()
```

## 커널밀도그림 (Kernel Density Plot)

- 히스토그램과 같은 연속적 분포를 곡선화해서 그린 그림
- sns.kdeplot()

```python
# in Histgram

x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)

plt.xticks(np.arange(0, 20, 2))

plt.hist(y, bins=np.arange(0, 22, 2))
plt.show()
```

```python
# kdeplot

sns.kdeplot(y, shade=True)
plt.show()
```

```python
vote_df = pd.DataFrame({"name":['Andy','Bob','Cat'], "vote":[True, True, False]})

vote_df
```

```python
# in matplotlib barplot

vote_count = vote_df.groupby('vote').count()
print(vote_count)

plt.bar(x=[False, True], height = vote_count['name'])
plt.show()
```

```python
# sns의 countplot

sns.countplot(x=vote_df['vote'])

plt.show()
```

## 캣 그림 (Cat Plot)

- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- `sns.catplot()`

```python
covid = pd.read_csv("./country_wise_latest.csv")

covid.head(5)
```

```python
s = sns.catplot(x='WHO Region', y='Confirmed', data=covid, kind='violin')
s.fig.set_size_inches(10, 6)

plt.show()
```

## 스트립 그림 (Strip Plot)

- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프
- `sns.stripplot()`

```python
s = sns.stripplot(x='WHO Region', y = 'Recovered', data=covid)

plt.show()
```

```python
#cf) swarmplot

s = sns.swarmplot(x='WHO Region', y = 'Recovered', data=covid)

plt.show()
```

## 히트맵 (Heatmap)

- 데이터의 행렬을 색상으로 표현해주는 그래프
- `sns.heatmap()`

```python
# 히트맵 예제

covid.corr()
```

```python
sns.heatmap(covid.corr())
plt.show()
```
